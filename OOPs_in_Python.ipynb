```python
# Four features in OOPs
# 1. Abstraction
# 2. Encapsulation
# 3. Inheritance
# 4. Polymorphism


# Abstraction - hiding unnecesary details from users through class, methods
# Encapsulation - restrict access to certain attributes or methods to protect data and enforce control access
# Inheritance - allows one class (child) to reuse the property and methods of another class (parent)
# Polymorphism - allow methods in diff classes to have same name but diff behaviour depending on object
```

```python
# Abstraction - hiding unnecesary details from users through class, methods

class Student:
  def __init__(self, name, grade, percentage):
    self.name = name
    self.grade = grade
    self.percentage = percentage


  def student_details(self): # method - abstraction
    print(f'{self.name} is in the class {self.grade}, with {self.percentage+2}%') # hidden from users - (jesy +2 add krra hua aur puri student detail line)


student1 = Student('Keshav', 12, 95)

student2 = Student('Wasiq', 9, 82)


student1.student_details()
student2.student_details()
```

```python
# Encapsulation - restrict access to certain attributes or methods to protect data and enforce control access

class Student:
  def __init__(self, name, grade, percentage):
    self.name = name
    self.grade = grade
    self.__percentage = percentage # double underscore limit access


  def get_percentage(self):  # best practice mei ap get word use krte jis chez ko ap ne encapsulate krra hw
    return self.__percentage


  def student_details(self): # method
    print(f'{self.name} is in the class {self.grade}, with {self.percentage}%')



student1 = Student('Keshav', 12, 95)

student2 = Student('Wasiq', 9, 82)


# student1.student_details()
# student2.student_details()
print(student1.get_percentage())
```

```python
# Inheritance - allows one class (child) to reuse the property and methods of another class (parent)


# parent class - baap
class Student:
  def __init__(self, name, grade, percentage):
    self.name = name
    self.grade = grade
    self.percentage = percentage


  def student_details(self): # method
    print(f'{self.name} is in the class {self.grade}, with {self.percentage}%')


# object - instance of class
student1 = Student('Keshav', 12, 95)
student2 = Student('Wasiq', 9, 82)


# child class - beta
class GraduateStudent(Student): # graduatestudent child class inherit prop and method from Student Parent Class
  def __init__(self, name, grade, percentage, stream): # old parameters from parents class and new parameter in child class
    super().__init__(name, grade, percentage) # call parent class init

    self.stream = stream # new attribute in child class


  def student_details(self):
    super().student_details() # method inherit from parent class
    print(f'Stream is {self.stream}')


# object
Grad_Student1 = GraduateStudent('Alishba', 12, 92, 'PCM')
# print(Grad_Student1.stream)

Grad_Student1.student_details()


```

```python
# Polymorphism - allows methods in diff classes to have same name but diff behaviour depending on object

# same method ka name diff class k andr diff tarike se behave krta hai

# parent class
class Student:
  def __init__(self, name, grade, percentage):
    self.name = name
    self.grade = grade
    self.percentage = percentage


  def student_details(self): # method
    print(f'{self.name} is in the class {self.grade}, with {self.percentage}%')


# object - instance of class
student1 = Student('Keshav', 12, 95)
student2 = Student('Wasiq', 9, 82)


# child class
class GraduateStudent(Student):
  def __init__(self, name, grade, percentage, stream):
    super().__init__(name, grade, percentage)
    self.stream = stream


  def student_details(self): # method
    # print(f'{self.name} is in class {self.grade}, with {self.percentage}% and from stream {self.stream}')
    print("Same method with diff output.")


# object - Student class
student1 = Student('Keshav', 12, 95)
student2 = Student('Wasiq', 9, 82)

# object - GraduateStudent class
Grad_Student1 = GraduateStudent('Alishba', 12, 92, 'PCM')


student1.student_details()
Grad_Student1.student_details()
```

**Decorators**

```python
# A function that extend the behaviour of another function.
# without modify base function
# Pass the base function as an argument to the decorator
```

```python
def add_sprinkles(func): # decorator
  def wrapper():
    print('**You add sprinkles**')
    func() # base function ko call karta hai
  return wrapper

@add_sprinkles
def get_icecream():  # base function
  print('Here is your icecream')

get_icecream()
```

```python
def add_sprinkles(func): # decorator
  def wrapper():
    print('**You add sprinkles**')
    func()
  return wrapper

def add_fudge(func): # decorator
  def wrapper():
    print('**You add fudge**')
    func()
  return wrapper

@add_sprinkles
@add_fudge
def get_icecream():  # base function
  print('Here is your icecream')

get_icecream()
```

```python
def add_sprinkles(func): # decorator
  def wrapper(*args, **kwargs):
    print('**You add sprinkles**')
    func(*args, **kwargs)
  return wrapper

def add_fudge(func): # decorator
  def wrapper(*args, **kwargs):
    print('**You add fudge**')
    func(*args, **kwargs)
  return wrapper

@add_sprinkles
@add_fudge
def get_icecream(flavor):  # base function
  print(f'Here is your {flavor} icecream')

get_icecream("vanila")
```
